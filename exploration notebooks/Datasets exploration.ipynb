## Genre

```python
import torch
from tgb.nodeproppred.dataset_pyg import PyGNodePropPredDataset
```

```python
dataset = PyGNodePropPredDataset(name='tgbn-genre', root="datasets")
```

```python
data = dataset.get_TemporalData()
```

```python
#Random sampling of 100 users
subset = torch.randperm(len(data.src.unique()))[:100]
```

```python
# Filtering the data with a mask
mask=torch.isin(data.src, subset)

filtered_src = data.src[mask]
filtered_dst = data.dst[mask]
filtered_t = data.t[mask]
filtered_msg = data.msg[mask]
```
